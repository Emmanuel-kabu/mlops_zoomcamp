```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import IsolationForest
from sklearn.feature_extraction import DictVectorizer
from sklearn.linear_model import Lasso
from sklearn.linear_model import Ridge
from sklearn.metrics import root_mean_squared_error
import numpy as np
```

```python
iso = IsolationForest(contamination=0.01, random_state=42)
yellow_trip_data["outlier"] = iso.fit_predict(yellow_trip_data.select_dtypes(include = [np.number]))
print(f"percentage of outliers: {yellow_trip_data['outlier'].value_counts(normalize=True)[1]*100}")
```

```python
yellow_trip_data["duration"] = yellow_trip_data["tpep_dropoff_datetime"] - yellow_trip_data["tpep_pickup_datetime"]
#yellow_trip_data["duration"] = yellow_trip_data["duration"].dt.total_seconds()
yellow_trip_data.duration = yellow_trip_data.duration.apply(lambda td: td.total_seconds() / 60)
```

```python
print(yellow_trip_data["duration"].describe())
```

```python

encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore')

# Select categorical columns only
categorical_cols = yellow_trip_data.select_dtypes(include=['object']).columns

# Fit-transform them
one_hot = encoder.fit_transform(yellow_trip_data[categorical_cols].astype(str))

# Optionally convert result to DataFrame
encoded_df = pd.DataFrame(one_hot, columns=encoder.get_feature_names_out(categorical_cols))

# Now you can concatenate it back to your original DataFrame
yellow_trip_data_encoded = pd.concat([yellow_trip_data.drop(columns=categorical_cols), encoded_df], axis=1)
```

```python
cat_cols = ['PULocationID', 'DOLocationID']
```

```python
#train_dict = yellow_trip_data_encoded[cat_cos].to_dict(orient='records')
yellow_trip_data_encoded[cat_cols] =  yellow_trip_data_encoded[cat_cols].astype(str)
train_dict =yellow_trip_data_encoded[cat_cols].to_dict(orient='records')
dv = DictVectorizer()
X_train = dv.fit_transform(train_dict)
```

```python
X_train.shape
```

```python
y_train = yellow_trip_data["duration"].values

lr = LinearRegression()
lr.fit(X_train, y_train)

y_pred = lr.predict(X_train)

error = root_mean_squared_error(y_train, y_pred)
```
